```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
# Load MNIST dataset
mnist = fetch_openml('mnist_784')
x = mnist.data  # Shape: (70000, 784)
y = mnist.target.astype(int)  # Labels (0-9)

# Convert to DataFrame for nice display
df = pd.DataFrame(x)
df['label'] = y

# Show first 5 rows
print(df.head())
# Select a subset of 6000 samples
x_subset = x[:6000]
y_subset = y[:6000]

# Normalize the data
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(x_subset)
```

```python
# Compile and train
model.compile(optimizer='adam', loss='mse')
model.fit(x_normalized, # input
          x_normalized, # target (output should be same as input)
          epochs=50, # no.of times to iterate over data
          batch_size=128, # number of samples per gradient update
          verbose=1) # show progress bar
```

```python
import matplotlib.pyplot as plt
# Get 3D representation of the data
Z = model.predict(x_normalized)

# 3D Scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(Z[:, 0], Z[:, 1], Z[:, 2], c=y_subset, cmap='jet', alpha=0.6)

# Labels and title
ax.set_title("3D Scatter Plot of MNIST (Reduced to 3 Features)")
ax.set_xlabel("Z Dimension 1")
ax.set_ylabel("Z Dimension 2")
ax.set_zlabel("Z Dimension 3")
plt.colorbar(scatter, ax=ax, label="Digit Labels")
plt.show()
```
